# The Graphical Method (Unbounded Solutions)

(from Nagraj Chapter 2) When an LP model has a bounded feasible region, as in the Flair Furniture example (i.e., it has an explicit boundary in every direction), it has an identifiable optimal corner point solution. 

However, if the feasible region is unbounded in one or more directions, as in the Holiday Meal Turkey Ranch example, depending on the objective function, the model may or may not have a finite solution. 

In the Holiday Meal problem, for example, we were able to identify a finite solution because the optimal corner point existed on the bounded side (see below). However, what happens if the objective function is such that we have to move our level profit (or cost) lines away from the bounded side? 

This commonly happens when you choose a improper objective function or forget to include a constraint in your model.

Now suppose that instead of minimizing cost, the owner of the ranch wants to use a different objective function. Specifically, based on his experience with the feeds and their fattening impact on his turkeys, assume that the owner estimates that brand A feed yields a **“ fattening value”** of 8 per pound, while brand B feed yields a fattening value of 12 per pound. The owner wants to find the diet that maximizes the total fattening value while ignoring cost.

So we update our objective function to be $Max(Z) = 8A + 12B$

But uh oh! There is a problem... the problem looks the same as before, the solution will be in the feasible region, but profit gets bigger as we move in the unbounded direction (if we keep feeding our chickens, they will get infinitely fat and we will be infinitely rich!)

## Some points for the isoline
We know the equation of the line (it's just the objective function... Z = 8A + 12B)

Isoline 1
* When A = 10, B = 0, Z = 10*8 = 80 (a nice easy point on the axis)
* When A = 0, we can solve for B
  * 12B = 80 so B = 80/12 = 6.667

Isoline 2
* When A = 12, B = 0, Z = 8*12 = 96 (a nice easy point on the axis)
* When A = 0, we can solve for B
  * 12B = 96 so B = 96/12 = 8



```
# clean plotting
%matplotlib inline
from pylab import *

figure(figsize=(6,6))
subplot(111, aspect='equal')
axis([0,10,0,10])
xlabel('Pounds of A')
ylabel('Pounds of B')

# Protein Constraint
# can you see the two points (0, 4.5) and (9, 0) in the code below?
# remember, A is X and B is Y 
x = array([0,   9])
y = array([4.5, 0]) 
plot(x,y,'r',lw=2)
# since this is a minimization problem we shade towards the origin!
# again, notice how the points line up!
fill_between([0,9,0], # x area
             [4.5, 0, 0], # y area
             [0,0,0], # bottom area, shade in everything below!
             color='red', # color 
             alpha=0.15) # transparency

# # Vitamin Constraint
x = array([0, 6])
y = array([8, 0]) 
plot(x,y,'blue',lw=2)
fill_between([0, 6, 0], # x area
             [8,0,0], # y area
             [0,0,0], # bottom area - all 0s! Easy.
             color='blue', # color 
             alpha=0.15) # transparency

# Iron constraint
# Iron is only in A! 
# the points are where it intersects the X axis (A = 3, B = 0)
# and the other point goes to infinity (we set to plot limits) (A = 3, B = 10)
# then draw a line between the two points
plot([3, 3],[0,10],'black',lw=2)
fill_between([0,3], # A values (X)
             [10, 10], # B values (Y)
             [0,0], # shade to left to the origin!
             color='grey',alpha=0.15)

legend(['Protein Constraint','Vitamin Constraint','Iron Constraint'])

# point 1
plot(3,4,'r.',ms=20) # makes a red dot on the plot
annotate('1', xy=(3,4), xytext=(3,4)) # adds a label for '1'

# point 2
plot(4.2, 2.4,'r.',ms=20) # makes a red dot on the plot
annotate('2', xy=(4.2, 2.4), xytext=(4.2, 2.4)) # adds a label for '2'

# point 3
plot(9,0,'r.',ms=20) # makes a red dot on the plot
annotate('3', xy=(9,0), xytext=(9,0)) # adds a label for '3'

# infinite profit!
# remember, the objective function is a level line... 
# let's add some examples

# level line 1
x = array([0, 10])
y = array([6.667, 0]) 
plot(x,y,'gray',lw=2, linestyle='dashed')
text(3.5, 3.5, '8A + 12B = 80', rotation=-35)

# level line 2
x = array([0, 12])
y = array([8, 0]) 
plot(x,y,'gray',lw=2, linestyle='dashed')
text(4.3, 4.3, '8A + 12B = 96', rotation=-35)


# add some bold text to the plot
text(5.5,5.5,'Increasing Profit', rotation=-35, fontweight='bold')
text(5.5,5.51,'Unbounded Feasible Region!', rotation=-35, fontweight='bold')
# add arrows
annotate('', xy=(5,7.5), xytext=(3.5,5.5),
          arrowprops=dict(width=0.5,headwidth=5))
annotate('', xy=(5,8.5), xytext=(3.5,6.5),
          arrowprops=dict(width=0.5,headwidth=5))
```

# What did we learn?
Be careful when setting up your linear program - are your chickens getting infinitey fat such that you'll make infinite money?!

Make sure your objective function makes sense and that you have included all of your constraints (and that the inequalities are pointing in the correct direction).
